                         Create a BAR Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ['Python','Java','C','C++']          
y = [85,70,60,40]
plt.bar(x,y)
```

```python
plt.title("BAR Plot")
plt.xlabel("Subject")
plt.ylabel("Number")
plt.bar(x,y)
```

```python
plt.title("BAR Plot")
plt.xlabel("Subject")
plt.ylabel("Number")
c= ['r','y','b','g']
plt.bar(x,y,width=0.5,color=c)
```

```python
plt.title("BAR Plot")
plt.xlabel("Subject")
plt.ylabel("Number")
c=['r','y','b','g']
plt.bar(x,y,width=0.5,color=c,align='edge',edgecolor = 'b',linewidth=5,linestyle='--',alpha=0.9,label="range")
plt.legend()
```

```python
x = ['Python','Java','C','C++']          
y = [85,70,60,40]
z = [20,30,40,50]
p = np.arange(len(x))
width = 0.2
p1 = [j + width for j in p]

plt.title("BAR Plot")
plt.xlabel("Subject",fontsize=15)
plt.ylabel("Number",fontsize=15)

plt.bar(p,y,width,color='r',label="Popularity")
plt.bar(p,z,width,color='y',label="Popularity1")
plt.legend()
```

```python
x = ['Python','Java','C','C++']          
y = [85,70,60,40]
z = [20,30,40,50]
p = np.arange(len(x))
width = 0.2
p1 = [j + width for j in p]

plt.title("BAR Plot")
plt.xlabel("Subject",fontsize=15)
plt.ylabel("Number",fontsize=15)

plt.bar(p,y,width,color='r',label="Popularity")
plt.bar(p1,z,width,color='y',label="Popularity1")
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
```

                                      Pie Plot

```python
x = [10,20,30,40]

plt.pie(x)
plt.show()
```

```python
x = [10,20,30,40]
y = ('C','C++',"Java","Python")
ex =[0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)
plt.title("Pie plot")
plt.show()
```

```python
x = [10,20,30,40]
y = ('C','C++',"Java","Python")
ex =[0.3,0.0,0.0,0.0]
c = ["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.4,labeldistance=1.1)
plt.show()
```

```python
x = [10,20,30,40]
y = ('C','C++',"Java","Python")
ex =[0.3,0.0,0.0,0.0]
c = ["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.1,labeldistance=1.1,startangle=0
         , textprops={"fontsize":15},counterclock = False,wedgeprops={"edgecolor":"m","linewidth":2}
        , rotatelabels =False)
plt.legend(loc=2)
plt.title("Pie Plot")
plt.xlabel("python")
plt.ylabel("number")
plt.show()
```

                                        Stem Plot

```python
x = [10,20,30,40,50]
y = [1,2,3,4,5]
plt.xlabel("Base")
plt.ylabel("Points")
plt.title("Stem Plot")
plt.stem(x,y,markerfmt = ".",linefmt = "--")
plt.show()
```

```python
x = [10,20,30,40,50]
y = [1,2,3,4,5]
plt.xlabel("Base")
plt.ylabel("Points")
plt.title("Stem Plot")
plt.stem(x,y,markerfmt = "or",linefmt = "--",bottom=0,basefmt="g")
plt.show()
```

```python
x = [10,20,30,40,50]
y = [1,2,3,4,5]

plt.stem(x,y,markerfmt = "or",linefmt = "--",bottom=0,basefmt="g",label="Python",orientation = "horizontal")
plt.legend()
plt.show()
```

                                            Box Plot  & Whisker Plot

                                    stack Plot

```python
x = [1,2,3,4,5]
area  = [2,3,2,5,4]
plt.stackplot(x,area)
plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [3,5,2,6,4]
area3 = [1,3,5,7,4]
plt.title("Stack Plot")
l =["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','y','m'])
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [3,5,2,6,4]
area3 = [1,3,5,7,4]
plt.title("Stack Plot")
l =["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','y','m']
             , baseline ='sym')
plt.grid()
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [3,5,2,6,4]
area3 = [1,3,5,7,4]
plt.xlabel("x-Axis")
plt.ylabel("y-Axis")

plt.title("Stack Plot")
l =["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','y','m']
             , baseline ='wiggle')
plt.legend()
plt.show()
```

                        Fill _ Between Plot

```python
import numpy as np
```

```python
x =np.array([1,2,3,4,5])
y =np.array([1,2,3,4,5])
plt.plot(x,y)
plt.fill_between(x,y)
plt.show()
```

```python
x =np.array([1,2,3,4,5])
y =np.array([1,2,3,4,5])
plt.plot(x,y,color='r')
plt.fill_between(x=[2,4],y1=2,y2=4)

plt.show()
```

```python
x =np.array([1,2,3,4,5])
y =np.array([1,2,3,4,5])
plt.plot(x,y,color='r')
plt.title("Fill_Between plot")
plt.fill_between(x,y,color='m',where=(x>=2)&(x<=4))

plt.show()
```
